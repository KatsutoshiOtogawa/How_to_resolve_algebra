```python
from sympy import *

x = Symbol('x')

x**2 + 6*x
```

```python
eq = Eq(x**2 + 6*x,-9)
eq
```

```python
# solve 
solve(eq,x)
```

```python
# factor is factorization function
fx = x**2 + 8*x + 16
factorization = factor(fx)
factorization
```

```python
# expand is expression function
expand(factorization)
```

```python

# plot question.
# function f(x) (a <= x <= b) is explaining below.
# plot(f(x),(x,a,b))
plot(fx,(x))
```

```python
# cross function
gx = -2*x + 5
plot(fx,gx,(x))
```

```python
# solve intersection
eq = Eq(fx,gx)
solve(eq)
```
