```python
from sympy import *

x = Symbol('x')

x**2 + 6*x
```

```python
fx = Eq(x**2 + 6*x,-9)
fx
```

```python
collect(fx)
```

```python
# solve 
solve(fx,x)
```

```python
# factor is factorization function
factorization = factor(fx)
factorization
```

```python
# expand is expression function
expand(factorization)
```

```python

# plot question.
# function f(x) (a <= x <= b) is explaining below.
# plot(f(x),(x,a,b))
plot(fx,(x))
```
